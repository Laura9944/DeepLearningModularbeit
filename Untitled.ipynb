```python
import numpy as np
import cv2
import os
from tqdm import tqdm
```

# Datensatz erstellen

```python
mean = 0.0   # some constant
std = 1.0    # some constant (standard deviation)

images_without_noise = []
images_with_noise = []

for filepath in tqdm(os.listdir('COCO/')):
    img_bgr = cv2.imread('COCO/{0}'.format(filepath))
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    crop_img = img_rgb[0:145, 0:145]
    cv2.imwrite('COCO_crop/'+ filepath, crop_img) 
    images_without_noise.append(crop_img)
    
    noise_image = crop_img + np.random.normal(mean, std, crop_img.shape)*(255/4)
    noise_image = np.clip(noise_image, 0, 255).astype(int)
    images_with_noise.append(noise_image)
    cv2.imwrite('COCO_noise/'+ filepath, noise_image) 
    
images_without_noise = np.asarray(images_without_noise)
images_with_noise = np.asarray(images_with_noise)
```

```python
from matplotlib import pyplot as plt

plt.imshow(images_without_noise[100])
plt.show()
plt.imshow(images_with_noise[100])
plt.show()
```

# Datensatz laden

```python
#load Datasets
images_without_noise = []

for filepath in tqdm(os.listdir('COCO_crop/')):
    image = cv2.imread('COCO_crop/{0}'.format(filepath))
    image = image/255
    images_without_noise.append(image)#.reshape((100,100,3)))
    
images_with_noise = []

for filepath in tqdm(os.listdir('COCO_noise/')):
    image = cv2.imread('COCO_noise/{0}'.format(filepath))
    image = image/255
    images_with_noise.append(image)#.reshape((100,100,3)))
    
images_without_noise = np.asarray(images_without_noise)
images_with_noise = np.asarray(images_with_noise)
```

```python
from matplotlib import pyplot as plt

plt.imshow(images_without_noise[0])
plt.show()
plt.imshow(images_with_noise[0])
plt.show()
```

```python
#split 80/20
split_index = int(len(images_with_noise)*0.8)

x_val = images_with_noise[split_index:]
y_val = images_without_noise[split_index:]
x_train = images_with_noise[:split_index]
y_train = images_without_noise[:split_index]
y_train.shape
```

```python
from tensorflow.keras.layers import Conv2DTranspose, Conv2D, Input
import tensorflow as tf

class Net(tf.keras.Model): 
    def __init__(self):

        super(Net, self).__init__() 

        self.encoder = tf.keras.Sequential([ 
          Input(shape=(145, 145, 3)), 
          Conv2D(16, (3,3), activation='relu', padding='same', strides=2), 
          Conv2D(8, (3,3), activation='relu', padding='same', strides=2) ,
          Conv2DTranspose(8, kernel_size=3, strides=2, activation='relu', padding='same'), 
          Conv2DTranspose(16, kernel_size=3, strides=2, activation='relu', padding='same'), 
          Conv2D(3, kernel_size=(3,3), activation='sigmoid', padding='same') #hier einfacher mit werten von 0-1 und sigmoid
        ]) 
  
    def call(self, x): 
        encoded = self.encoder(x) 
        #decoded = self.decoder(encoded) 
        return encoded
```

```python
model = Net()
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x_train, y_train, 
                epochs=15,
                #shuffle=True, 
                validation_data=(x_val, y_val))
```
