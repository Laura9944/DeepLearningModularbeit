```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt


# only load images from set
(y_train, _), (y_test, _) = cifar10.load_data()
```

```python
#normalize image for computing speed
y_train = y_train.astype('float32') / 255.
y_test = y_test.astype('float32') / 255.
```

```python
#add noise
noise_scale = 0.1 
x_train = y_train + noise_scale * tf.random.normal(shape=y_train.shape) 
x_test = y_test + noise_scale * tf.random.normal(shape=y_test.shape)
x_train = tf.clip_by_value(x_train, clip_value_min=0., clip_value_max=1.) 
x_test = tf.clip_by_value(x_test, clip_value_min=0., clip_value_max=1.)
```

```python
#show some images
plt.imshow(tf.squeeze(y_test[0])) 
plt.show()
plt.imshow(tf.squeeze(x_test[0])) 
plt.show()
plt.imshow(tf.squeeze(y_test[10])) 
plt.show() 
plt.imshow(tf.squeeze(x_test[10]))
plt.show()
```

```python
from tensorflow.keras.layers import Conv2DTranspose, Conv2D, Input

class Net(tf.keras.Model): 
    def __init__(self):

        super(Net, self).__init__() 

        self.encoder = tf.keras.Sequential([ 
          Input(shape=(32, 32, 3)), 
          Conv2D(32, (3,3), activation='relu', padding='same', strides=2), 
          Conv2D(16, (3,3), activation='relu', padding='same', strides=2)]) 
    
        self.decoder = tf.keras.Sequential([ 
          Conv2DTranspose(16, kernel_size=3, strides=2, activation='relu', padding='same'), 
          Conv2DTranspose(32, kernel_size=3, strides=2, activation='relu', padding='same'), 
          Conv2D(3, kernel_size=(3,3), activation='sigmoid', padding='same')]) 
  
    def call(self, x): 
        encoded = self.encoder(x) 
        decoded = self.decoder(encoded) 
        return decoded
```

```python
autoencoder = Net()
```

```python
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
autoencoder.fit(x_train, 
                y_train, 
                epochs=10, 
                shuffle=True, 
                validation_data=(x_test, y_test))
```

```python
encoded_imgs=autoencoder.encoder(x_test).numpy()
decoded_imgs=autoencoder.decoder(encoded_imgs)
```

```python
n = 10 
ax = plt.subplot(2, 5, i + 1) 
plt.figure(figsize=(20, 7))
plt.gray()
for i in range(n): 
    # display original + noise 
    bx = plt.subplot(3, n, i + 1) 
    plt.title("original + noise") 
    plt.imshow(tf.squeeze(x_test[i])) 
    ax.get_xaxis().set_visible(False) 
    ax.get_yaxis().set_visible(False) 
  
    # display reconstruction 
    cx = plt.subplot(3, n, i + n + 1) 
    plt.title("reconstructed") 
    plt.imshow(tf.squeeze(decoded_imgs[i])) 
    bx.get_xaxis().set_visible(False) 
    bx.get_yaxis().set_visible(False) 
  
    # display original 
    ax = plt.subplot(3, n, i + 2*n + 1) 
    plt.title("original") 
    plt.imshow(tf.squeeze(y_test[i])) 
    ax.get_xaxis().set_visible(False) 
    ax.get_yaxis().set_visible(False) 

plt.show()
```
